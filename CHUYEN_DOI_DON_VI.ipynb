```python
import numpy as np
import pandas as pd

```

```python
# Tải tệp dữ liệu từ đường dẫn
data = pd.read_excel("/content/average_data.xlsx")

# Hiển thị dữ liệu
print(data.head(100))
```

```python
data.isnull()
```

```python
# Hệ số chuyển đổi
conversion_factor =  86400

# Lấy cột chứa dữ liệu lượng mưa (đơn vị: kg m^-2 s^-1)
rainfall_data_kgs = data["pr"].values

# Chuyển đổi đơn vị
rainfall_data_mm_per_day = rainfall_data_kgs * conversion_factor

# In kết quả
print("Dữ liệu lượng mưa ban đầu (kg m^-2 s^-1):")
print(rainfall_data_kgs)

# Chuyển đổi đơn vị và in ra 100 dòng
for i in range(1000):
    rainfall_kgs = rainfall_data_kgs[i]
    rainfall_mm_per_day = rainfall_kgs * conversion_factor
    print(f"Row {i+1}: {rainfall_kgs} kg m^-2 s^-1 = {rainfall_mm_per_day} mm/day")
```

```python
# Dữ liệu lượng mưa đã chuyển đổi (mm/day)
rainfall_data_mm_per_day = rainfall_data_kgs * conversion_factor

# Tạo DataFrame từ dữ liệu chuyển đổi
converted_data = pd.DataFrame({'Rainfall (mm/day)': rainfall_data_mm_per_day})

# Lưu dữ liệu vào file Excel
converted_data.to_excel('Rainfall_data_converted.xlsx', index=False)
```
